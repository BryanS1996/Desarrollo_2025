```python
# cargamos las librerias
import requests
import pandas as pd
import stanza
stanza.download('la')
from bs4 import BeautifulSoup as bs
```

```python
# Cargar modelo de análisis lingüístico en latín
nlp = stanza.Pipeline('la', processors='tokenize,pos,lemma')

url = "https://en.wikipedia.org/wiki/List_of_Latin_phrases_(full)"
tablas = pd.read_html(url)
df_todo = pd.concat(tablas, ignore_index=True)

# Guardar archivo Excel (opcional)
df_todo.to_excel("archivo.xlsx")

# Unir todo el texto
texto = ' '.join(df_todo.astype(str).values.flatten())

# --- Procesar el texto con Stanza ---
doc = nlp(texto)

# Extraer verbos y palabras
from collections import Counter

verbos = []
palabras = []

for sentence in doc.sentences:
    for word in sentence.words:
        if word.lemma:  # Asegura que hay lema
            lema = word.lemma.lower()
            palabras.append(lema)
            if word.upos == 'VERB':
                verbos.append(lema)

# Contar las más comunes
frecuencia_verbos = Counter(verbos).most_common(10)
frecuencia_palabras = Counter(palabras).most_common(10)

# Mostrar resultados
print("🔹 Verbos más frecuentes en latín:")
for verbo, freq in frecuencia_verbos:
    print(f"{verbo}: {freq}")

print("\n🔹 Palabras más frecuentes en latín:")
for palabra, freq in frecuencia_palabras:
    print(f"{palabra}: {freq}")

# --- Guardar resultados en Excel ---
df_resultado = pd.DataFrame({
    'Verbos': [v[0] for v in frecuencia_verbos],
    'Frecuencia verbo': [v[1] for v in frecuencia_verbos],
    'Palabras': [w[0] for w in frecuencia_palabras],
    'Frecuencia palabra': [w[1] for w in frecuencia_palabras]
})

df_resultado.to_excel('palabras_latinas.xlsx')
```
